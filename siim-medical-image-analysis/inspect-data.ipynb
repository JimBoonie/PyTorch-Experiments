```python
import os
import numpy as np
from imageio import imread
import matplotlib.pyplot as plt
%matplotlib inline

data_path = '/home/masonmcgough/Workspace/Data/siim-medical-image-analysis-tutorial/tiff_images'
VALID_EXTS = ('.jpg', '.png', '.tif', '.tiff')
```

```python
file_paths = []
for e in VALID_EXTS:
    file_paths.extend([
        os.path.join(data_path, f) 
        for f in os.listdir(data_path) 
        if f.endswith(e)
    ])
    
print("{0} files found in '{1}'".format(len(file_paths), data_path))
print("file_paths[0]: '{0}'".format(file_paths[0]))
```

## Load one image

```python
2**12
```

```python
# Compute range of one image in set
img_name = file_paths[0]
img = imread(img_name)

print("Img 0:")
print("\tDimensions: {0}".format(img.shape))
print("\tRange: [{0}, {1}]".format(np.min(img), np.max(img)))

# Compute range of whole set
set_min = float('inf')
set_max = float('-inf')
for f in file_paths:
    f_img = imread(f)
    set_min = min(np.min(f_img), set_min)
    set_max = max(np.max(f_img), set_max)
    
print("Whole Set:")
print("\tRange: [{0}, {1}]".format(set_min, set_max))
```

## Display image

```python
# let's assume the range is [-2**12, 2**12]
img = (img + 2**12) / 2**13
print("New range: [{0}, {1}]".format(np.min(img), np.max(img)))
img_plot = plt.imshow(img)
```
